# Data cleaning:

```python
df_new= df[['E','SV','SHO','R']]
```

we found that ['E','SV','SHO','R']   there are the columns from which  we have to remove outlier as well as skewness as outlier and skew we should not remove from label

```python
df_new.shape
```

```python
df_new.columns
```

```python
from scipy.stats import  zscore
z=np.abs(zscore(df_new))
z

dfnew=df_new[(z<3).all(axis=1)]
```

```python
dfnew.shape
```

```python
Dataloss=((30-29)/30)*100
Dataloss
```

Data loss is 3% so we  remove outlier as it is not more than 8%

```python
dfnew.skew()
```

```python
from sklearn.preprocessing import power_transform
```

```python
dfnew['E'] = power_transform(np.array(dfnew['E']).reshape(-1,1), method='yeo-johnson')
```

```python
dfnew['SV'] = power_transform(np.array(dfnew['SV']).reshape(-1,1), method='yeo-johnson')
```

```python
dfnew['SHO'] = power_transform(np.array(dfnew['SHO']).reshape(-1,1), method='yeo-johnson')
```

```python
dfnew['R'] = power_transform(np.array(dfnew['R']).reshape(-1,1), method='yeo-johnson')
```

```python
dfnew.skew()
```

```python
dfnew[['W', 'AB', 'H', '2B', '3B', 'HR', 'BB', 'SO', 'SB', 'RA', 'ER',
       'ERA', 'CG', ]]=df[['W', 'AB', 'H', '2B', '3B', 'HR', 'BB', 'SO', 'SB', 'RA', 'ER',
       'ERA', 'CG', ]]
```

```python
dfnew
```

Data preprocessing:

```python
y = dfnew['W']
```

```python
x=dfnew.drop(('W'), axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
import  statistics as  st

from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x, y,test_size=.22 , random_state= 2)
```

```python
from  sklearn.linear_model  import LinearRegression
from sklearn.metrics  import    mean_squared_error ,r2_score
```

```python
lg=LinearRegression()
lg.fit(x_train,y_train)
lg.score(x_train,y_train)
```

```python
from sklearn.linear_model   import Lasso,Ridge
```

```python
ls=Lasso(0.001)
ls.fit(x_train,y_train)
ls.score(x_train,y_train)
```

```python
rd=Ridge(0.001)
rd.fit(x_train,y_train)
rd.score(x_train,y_train)
```

```python
from sklearn.tree  import DecisionTreeRegressor

from  sklearn.svm import SVR

from sklearn.ensemble import RandomForestRegressor
from  sklearn.neighbors  import KNeighborsRegressor
```

```python
kmn=KNeighborsRegressor()
kmn.fit(x_train,y_train)
predkmn=kmn.predict(x_test)
print("mse==",mean_squared_error(y_test,predkmn))
print("R2SCORE" ,r2_score(y_test,predkmn))
```

# Observation--here model is overfitting so we will be hypertuning model to overcome this uunderfit issue

```python
rvr=RandomForestRegressor()
rvr.fit(x_train,y_train)
predrvr=rvr.predict(x_test)
print("mse==",mean_squared_error(y_test,predrvr))
print("R2SCORE" ,r2_score(y_test,predrvr))
```

```python
dtr=DecisionTreeRegressor()
dtr.fit(x_train,y_train)
preddtc=dtr.predict(x_test)
print("mse==",mean_squared_error(y_test,preddtc))
print("R2SCORE" ,r2_score(y_test,preddtc))
```

```python
from xgboost  import XGBRegressor
xg=XGBRegressor()
xg.fit(x_train,y_train)
preddtc=xg.predict(x_test)
print("mse==",mean_squared_error(y_test,preddtc))
print("R2SCORE" ,r2_score(y_test,preddtc))
```

```python
from sklearn.ensemble import AdaBoostRegressor

ad=AdaBoostRegressor()
ad.fit(x_train,y_train)
predrvr=ad.predict(x_test)
print("mse==",mean_squared_error(y_test,predrvr))
print("R2SCORE" ,r2_score(y_test,predrvr))
```

# Now we are calculating cross val score

# As model is overfitting having huge difference between train and test score so we will try to predict best model through hyper tuning

```python
from sklearn.model_selection import cross_val_score
from sklearn.datasets import make_blobs


x, y = make_blobs(n_samples=10000, n_features=10, centers=100,random_state=0)

clf = DecisionTreeRegressor(max_depth=None, min_samples_split=2,random_state=0)
scores = cross_val_score(clf, x, y, cv=5)
scores.mean()



```

```python


clf = RandomForestRegressor(n_estimators=10, max_depth=None,min_samples_split=2, random_state=0)
scores = cross_val_score(clf, x, y, cv=5)
scores.mean()


```

```python
clf = XGBRegressor(n_estimators=10, max_depth=None, random_state=0)
scores = cross_val_score(clf, x, y, cv=5)
scores.mean() 
```

```python
clf = AdaBoostRegressor(n_estimators=10, random_state=0)
scores = cross_val_score(clf, x, y, cv=5)
scores.mean() 
```

From  above we found that Randomforestregressor is best model
