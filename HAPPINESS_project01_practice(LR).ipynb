```python
import  statistics as  st

from sklearn.model_selection import train_test_split
```

```python
from  sklearn.linear_model   import LinearRegression
```

```python
x_train,x_test,y_train,y_test=train_test_split(x, y,test_size=.22 , random_state= 2)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
lg=LinearRegression()
```

```python
lg.fit(x_train,y_train)
```

```python
lg.score(x_train,y_train)
```

```python
lg.coef_
```

```python
lmpred=lg.predict(x_test)
```

```python
from sklearn.metrics  import    mean_squared_error ,r2_score
```

```python
print("mse==",mean_squared_error(y_test,lmpred))
print("R2SCORE" ,r2_score(y_test,lmpred))
```

```python
from  sklearn.svm import SVR
```

```python
from sklearn.tree  import DecisionTreeRegressor
```

```python
from  sklearn.neighbors  import KNeighborsRegressor
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
svr=SVR(kernel='rbf')
svr.fit(x_train,y_train)
predsvr=svr.predict(x_test)
print("mse==",mean_squared_error(y_test,predsvr))
print("R2SCORE" ,r2_score(y_test,predsvr))
```

```python
svr=SVR(kernel='poly')
svr.fit(x_train,y_train)
predsvr=svr.predict(x_test)
print("mse==",mean_squared_error(y_test,predsvr))
print("R2SCORE" ,r2_score(y_test,predsvr))
```

```python
dtr=DecisionTreeRegressor()
dtr.fit(x_train,y_train)
preddtc=dtr.predict(x_test)
print("mse==",mean_squared_error(y_test,preddtc))
print("R2SCORE" ,r2_score(y_test,preddtc))
```

```python
rvr=RandomForestRegressor()
rvr.fit(x_train,y_train)
predrvr=rvr.predict(x_test)
print("mse==",mean_squared_error(y_test,predrvr))
print("R2SCORE" ,r2_score(y_test,predrvr))
```

```python
from sklearn.model_selection  import cross_val_score
```

```python
print(cross_val_score(dtr,x,y,cv=5).mean())
```

```python
print(cross_val_score(rvr,x,y,cv=5).mean())
```

```python
print(cross_val_score(svr,x,y,cv=5).mean())
```

we found  difference  is less between DTR  model r2 score and coss validation  hence DTR is my best model
